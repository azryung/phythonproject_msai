```python
# 도시명을 입력하면 위도와 별자리 리스트를 반환하는 함수 만들기

def predict_best_meteor_shower_viewing(city):
    # 도시의 위도 정보 불러오기 (Series 형태로)
    latitude_series = cities.loc[cities['city'] == city, 'latitude']
    meteor_showers_string = ''
    # 입력한 도시 정보가 존재하지 않으면
    if latitude_series.empty:
        meteor_showers_string = f"{city}에서는 별자리 관측이 불가능합니다"
        return meteor_showers_string
    # 도시 정보가 존재하면
    else:
        # 위도 값 추출
        latitude = latitude_series.iloc[0]
        
        # 해당 도시의 위도에서 관측 가능한 별자리 조회
        constellations_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (constellations['latitudeend'] <= latitude), 'constellation'].tolist()
        
        meteor_showers_string = f"{city}에서는 별자리 관측이 가능합니다"
        
        # 도시 위도, 별자리 리스트, 메시지 반환
    return latitude, constellations_list, meteor_showers_string
```

## 최적의 유성우 관측일

```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기

def predict_best_meteor_shower_viewing(city):

    # 입력한 도시 정보가 존재하지 않으면 오류 메시지
    
    latitude_series = cities.loc[cities['city'] == city, 'latitude']
    meteor_showers_string = ''
    # 입력한 도시 정보가 존재하지 않으면
    if latitude_series.empty:
        meteor_showers_string = f"{city}에서는 별자리 관측이 어렵습니다. 다른 지역을 확인해보시겠어요?"
        return meteor_showers_string
    # 도시 정보가 존재하면 위도 정보를 불러온다
    else:
        # 위도 값 추출
        latitude = latitude_series.iloc[0]
        
        # 해당 도시의 위도에서 관측 가능한 별자리 조회
        constellations_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (constellations['latitudeend'] <= latitude), 'constellation'].tolist()
        cons_len = len(constellations_list)
        meteor_showers_string = f"{city}는 위도 {latitude} 위치합니다. 총 {cons_len}개의 별자리 관측이 가능합니다.\n별자리는 {constellations_list} 입니다.\n "


        #도시별 관찰 가능한 별다지 iterate for 문
        for cons in constellations_list :
            meteor_shower = meteor_showers.loc[meteor_showers['radiant'] == cons, 'name'].iloc[0]
            meteor_showers_startdate = meteor_showers.loc[meteor_showers['radiant'] == cons, 'startdate'].iloc[0]
            meteor_showers_enddate = meteor_showers.loc[meteor_showers['radiant'] == cons, 'enddate'].iloc[0]
 
            moon_phase_list = moon_phases.loc[(moon_phases['date'] >= meteor_showers_startdate) & (moon_phases['date'] <= meteor_showers_enddate)]
            bestdate=moon_phase_list.loc[moon_phase_list['Percentage'].idxmin()]['date']
            meteor_showers_string += f'\n{bestdate.to_pydatetime().strftime("%Y년 %m월 %d일")}에 {cons}별자리를 보시면 {meteor_shower} 유성우를 볼 수 있습니다.'
            final_comment = f"\n꿈꾸시던 유성우를 만날 수 있기를 응원할게요!"
        return meteor_showers_string + final_comment
        

  

    #     # 유성우가 보일 때의 달의 위상 조회


        

    #     # 달이 눈에 보이는 첫 날 조회
        

    #     # 사용자에게 정보전달
         


    # return meteor_shower_string
```

```python

def draw_stars():
    """10x10 별 그리드를 문자열로 반환합니다."""
    star_grid = ""
    for _ in range(10):
        star_grid += "⭐" * 10 + "\n"
    return star_grid







# 도시명을 입력히면 별자리 정보를 알려주는 함수 완성하기






def predict_best_meteor_shower_viewing(city):

    # 입력한 도시 정보가 존재하지 않으면 오류 메시지
    
    latitude_series = cities.loc[cities['city'] == city, 'latitude']
    meteor_showers_string = ''
    # 입력한 도시 정보가 존재하지 않으면
    if latitude_series.empty:
        meteor_showers_string = f"{city}에서는 별자리 관측이 어렵습니다. 다른 지역을 확인해보시겠어요?"
        return meteor_showers_string
    # 도시 정보가 존재하면 위도 정보를 불러온다
    else:
        # 위도 값 추출
        latitude = latitude_series.iloc[0]
        
        # 해당 도시의 위도에서 관측 가능한 별자리 조회
        new_constellations_list = new_constellations.loc[(new_constellations['latitudestart'] >= latitude) & (new_constellations['latitudeend'] <= latitude), 'constellation'].tolist()
        new_cons_len = len(new_constellations_list)
        meteor_showers_string = f"{city}는 위도 {latitude} 위치합니다. 총 {new_cons_len}개의 별자리 관측이 가능합니다.\n별자리는 {new_constellations_list} 입니다.\n "


        #도시별 관찰 가능한 별다지 iterate for 문
        for new_cons in new_constellations_list :
            new_meteor = new_meteor_showers.loc[new_meteor_showers['radiant'] == new_cons, 'name'].iloc[0]
            new_startdate = new_meteor_showers.loc[new_meteor_showers['radiant'] == new_cons, 'startdate'].iloc[0]
            new_enddate = new_meteor_showers.loc[new_meteor_showers['radiant'] == new_cons, 'enddate'].iloc[0]
 
            moon_phase_list = moon_phases.loc[(moon_phases['date'] >= new_startdate) & (moon_phases['date'] <= new_enddate)]
            bestdate=moon_phase_list.loc[moon_phase_list['Percentage'].idxmin()]['date']
            meteor_showers_string += f'\n{bestdate.to_pydatetime().strftime("%Y년 %m월 %d일")}에 {new_cons}별자리를 보시면 {new_meteor} 유성우를 볼 수 있습니다.'
            final_comment = f"\n꿈꾸시던 유성우를 만날 수 있기를 응원할게요!"
        return f"{meteor_showers_string}{final_comment}\n\n{draw_stars()}"
```

```python
def draw_stars():
    """
    for 반복문을 사용해서 픽셀 아트 스타일의 하트 모양을 문자열로 반환합니다.
    """
    heart_art = ""
    
    # 하트의 위쪽(두 덩어리) 그리기
    for i in range(4):
        heart_art += (' ' * (3 - i) + '*' * (2 * i + 1) + ' ' * (5 - 2 * i) + '*' * (3 * i + 1) + '\n')

    # 하트의 중간(합쳐지는 부분) 그리기
    heart_art += '*' * 15 + '\n'
    
    # 하트의 아래쪽(뾰족한 부분) 그리기
    for i in range(1, 6):
        heart_art += (' ' * i + '*' * (15 - 3 * i) + '\n')

    return heart_art
print(draw_stars())
```
